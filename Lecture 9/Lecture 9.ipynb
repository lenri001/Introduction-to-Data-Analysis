```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=[0, 1, 2, 3, 4, 5, 6, 7, 8],
    y=[0, 1, 3, 2, 4, 3, 4, 6, 5]
))


fig.add_trace(go.Scatter(
    x=[0, 1, 2, 3, 4, 5, 6, 7, 8],
    y=[0, 4, 5, 1, 2, 2, 3, 4, 2]
))
section_position = [0,2,5,6,8]  
section_colors = ['lime', 'yellow', 'deepskyblue', 'red']
dictlist = [dict(type="rect", # The sahpe for the sections should always be a rectangle 
            # x-reference is assigned to the x-values
            xref="x",
            # y-reference is assigned to the plot paper [0,1]
            yref="paper",
            x0= section_position[i], 
            y0=0, # bottom of the paper 
            x1= section_position[i + 1],
            y1=1, # top of the paper 
            fillcolor= section_colors[i],
            opacity=0.5,
            layer="below",
            line_width=0) for i in range(len(section_position) - 1)]
# Showing the arrow is a boolean statement 
# While add_trace can take a list or array as an input, add-annotation only takes a numerical input such as int or float
# To plot an arrow realtive to the traces (plot lines) xref = "x" and yref = "y"
# To plot an arrow realtive to the plot frame regarless of the traces xref = "paper" and yref = "paper"
def find_midpoint(start, finish):
        midpoint = (finish - start) / 2 # Creates a time delta
        midpoint += start # Add time delta to start point
        return midpoint
# Arrows on the top
top_arrow_names = ["Section 1", "Section 2", "Section 3", "Section 4"]
for i in range(len(section_position) - 1):
    fig.add_annotation(x=find_midpoint(section_position[i],section_position[i+1]), y=1,
                text= top_arrow_names[i],
                showarrow=True,
                arrowhead=1,
                xref="x",
                yref="paper")
#Arrows within the line
line_arrow_position_x = [2, 7]
line_arrow_position_y = [5, 6]
line_arrow_names = ['Red Top', 'Blue Top']
for i in range(len(line_arrow_position_x)):
    fig.add_annotation(x=line_arrow_position_x[i], y=line_arrow_position_y[i],
                text=line_arrow_names[i],
                showarrow=True,
                arrowhead=1,
                xref="x",
                yref="y")
fig.update_layout(showlegend=False,shapes = dictlist)

fig.show()
```

```python
def plot_graph_label(x_data, y_data, line_name, title, x_title, y_title, file_title, x_label, label_name, label_color, show_fig = False):
    # Create Dictionary for Shape if needed
    dictlist = [dict(type="rect",
            # x-reference is assigned to the x-values
            xref="x",
            # y-reference is assigned to the plot paper [0,1]
            yref="paper",
            x0= x_label[i],
            y0=0,
            x1=x_label[i + 1],
            y1=1,
            fillcolor=label_color[i],
            opacity=0.5,
            layer="below",
            line_width=0,) for i in range((len(x_label) - 1))]
    # Find Midpoint between two timestamps
    def find_midpoint(start, finish):
        midpoint = (finish - start) / 2 # Creates a time delta
        midpoint += start # Add time delta to start point
        return midpoint
    # Define the figure
    fig = go.Figure()
    # Add data to figure
    for i in range(0, len(x_data), 1):
        if i == 0:
            fig.add_trace(go.Scattergl(x=x_data[i], y=y_data[i], name= line_name[i], line_color='deepskyblue'))
        else: 
            fig.add_trace(go.Scattergl(x=x_data[i], y=y_data[i], name= line_name[i]))
    # Add annotations 
    for i in range(0,(len(x_label) - 1)):
        fig.add_annotation(
                x=find_midpoint(x_label[i], x_label[i + 1]),
                y=1,
                xref="x",
                yref="paper",
                text=label_name[i],
                showarrow=True,
                align="center",
                arrowhead=2,
                arrowsize=1,
                arrowwidth=2,
                arrowcolor="#636363",
                ax=20,
                ay=-30,
                bordercolor=label_color[i],
                borderwidth=2,
                borderpad=4,
                bgcolor=label_color[i],
                opacity=0.8
                )
    
    # Create Layout for the Html
    
    fig.update_layout(
        title=go.layout.Title(
            text= title,
            xref="paper",
            x=0.5
        ),
        xaxis=go.layout.XAxis(
            title=go.layout.xaxis.Title(
                text=x_title,
            ),
            rangeslider=dict(
                visible=True
            )
        ),
        yaxis=go.layout.YAxis(
            title=go.layout.yaxis.Title(
                text=y_title,
            )
        ),
        shapes = dictlist
    )
    if show_fig:
        fig.show()
    fig.write_html((file_title + '_with_rangeslider.html'), include_mathjax='cdn')
    fig.update_layout(
        title=go.layout.Title(
            text= title,
            xref="paper",
            x=0.5
        ),
        xaxis=go.layout.XAxis(
            title=go.layout.xaxis.Title(
                text=x_title,
            )
        ),
        yaxis=go.layout.YAxis(
            title=go.layout.yaxis.Title(
                text=y_title,
            )
        ),
        shapes = dictlist
    )
    fig.write_html((file_title + '_without_rangeslider.html'), include_mathjax='cdn')
       
    # Create Layout for SVG, PDF, and other Static Images
    fig = go.Figure()
    # Add data to figure
    for i in range(0, len(x_data), 1):
        if i == 0:
            fig.add_trace(go.Scatter(x=x_data[i], y=y_data[i], name= line_name[i], line_color='deepskyblue'))
        else: 
            fig.add_trace(go.Scatter(x=x_data[i], y=y_data[i], name= line_name[i]))
    # Add annotations 
    for i in range(0,(len(x_label) - 1)):
        fig.add_annotation(
                x=find_midpoint(x_label[i], x_label[i + 1]),
                y=1,
                xref="x",
                yref="paper",
                text=label_name[i],
                showarrow=True,
                font=dict(
                    family="Courier New, monospace",
                    size=16,
                    color= 'black' #"#ffffff"
                    ),
                align="center",
                arrowhead=2,
                arrowsize=1,
                arrowwidth=2,
                arrowcolor="#636363",
                ax=20,
                ay=-30,
                bordercolor=label_color[i],
                borderwidth=2,
                borderpad=4,
                bgcolor=label_color[i],
                opacity=0.8
                )        
    
    fig.update_layout(
        title=go.layout.Title(
            text= title,
            xref="paper",
            x=0.5
        ),
        xaxis=go.layout.XAxis(
            title=go.layout.xaxis.Title(
                text=x_title,
            ),
        ),
        yaxis=go.layout.YAxis(
            title=go.layout.yaxis.Title(
                text=y_title,
            )
        ),
        shapes = dictlist
    )
    fig.write_image((file_title + '.svg'))
    fig.write_image((file_title + '.pdf'))
    return
```
