Why do we use HDF5 in Pandas?
At this point the usefulness of the Pandas DataFrame is evident. It is an excellent way of organizing, manipulating, and plotting large amounts of data. However, how would I store a dataframe for later use efficientl? This is where HDF5 format.

The HDF5 gives us the benefit of:
* A free and open source format
* A format that mantains the dataFrame format when stored 
* A format that can quickly import large amounts of data

Pandas and HDF5

Write to a HDF5
https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.to_hdf.html

Read to a HDF5
https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_hdf.html

Creating an HDF5

```python
df_big.to_hdf('water_district_clean_data.h5', 'df', format='fixed', mode='w')
```

Reading an HDF5

```python
df_2 = pd.read_hdf("water_district_clean_data.h5", "df")
df_2
```

```python
df_2["Random"] = 1
df_2
```

Updating an HDF5

```python
df_2.to_hdf('water_district_clean_data.h5', 'df', format='fixed', mode='r+')
```

```python
df_3 = pd.read_hdf("water_district_clean_data.h5", "df")
df_3
```
